### Function to get stock price

```python
def get_stock_price(stock_code, max_retries=3, backoff_factor=2):
    url = f"https://finance.yahoo.com/quote/{stock_code}"
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.9",
        "Connection": "keep-alive",
        "Referer": "https://sg.finance.yahoo.com/"
    }

    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=10)

            if response.status_code == 200:
                soup = BeautifulSoup(response.text, "html.parser")
                price_element = soup.find("span", {"data-testid": "qsp-price"})
                
                if price_element:
                    return float(price_element.text.replace(',', ''))
                else:
                    return "Price not found."
            else:
                print(f"Attempt {attempt + 1}: Failed to fetch data (Status Code: {response.status_code})")
        except requests.RequestException as e:
            print(f"Attempt {attempt + 1}: Request failed due to {e}")

        # Wait before retrying
        time.sleep(backoff_factor ** attempt)

    return "Failed to fetch data after retries."
```

### Compute FINANCIAL RATIOS (Balance Sheet + Income File)

```python
# ------------------------------------------------------------------
# 2) Compute FINANCIAL RATIOS (Balance Sheet + Income File)
# ------------------------------------------------------------------
all_stocks_data = []

for reit in data_extracted:
    stock_code = reit["Stock Code"]
    stock_code_yf = f"{stock_code}.SI"
    
    # Read all the CSV file in company_data folder that was previously extracted
    balance_sheet_file = f"company_data/{stock_code_yf}_balance_sheet.csv"
    income_file = f"company_data/{stock_code_yf}_income_statement.csv"     
    company_info_file = f"company_data/{stock_code_yf}_company_info.csv"
    dividends_file = f"company_data/{stock_code_yf}_dividends.csv"
    stock_price_file = f"company_data/{stock_code_yf}_stock_price_history_non_adj.csv"

    if not os.path.exists(balance_sheet_file) or not os.path.exists(company_info_file):
        print(f"Error!! Missing essential files for {stock_code_yf}, skipping ratio calculations...")
        continue

    # --- Load DataFrames ---
    df_balance = pd.read_csv(balance_sheet_file)
    df_income = pd.read_csv(income_file)
    df_info = pd.read_csv(company_info_file)
    df_dividends = pd.read_csv(dividends_file) if os.path.exists(dividends_file) else None

    # --- Helper to locate rows by regex in each sheet ---
    def find_balance_row(label_regex):
        return df_balance[df_balance.iloc[:, 0].str.contains(label_regex, na=False, case=False)]

    def find_income_row(label_regex):
        if df_income is not None:
            return df_income[df_income.iloc[:, 0].str.contains(label_regex, na=False, case=False)]
        # Return empty if no income file
        return pd.DataFrame()

    # --- Identify rows from the Balance Sheet ---
    total_assets_row = find_balance_row("Total Assets")
    total_debt_row = find_balance_row("Total Debt")
    total_liab_row = find_balance_row("Total Liabilities")
    equity_row = find_balance_row("Total Equity|Unitholders' Funds")
    shares_row = find_balance_row("Share Issued|Ordinary Shares Number")

    # --- Identify rows from the Income File (NPI, Interest Expense, etc.) ---
    npi_row = find_income_row("Net Income")
    interest_expense_row = find_income_row("Interest Expense")
    ebitda_row = find_income_row("EBITDA")

    # The columns for the balance sheet might be e.g. [Item, 2024-03-31, 2023-03-31,...]
    # We will assume the income file uses the same set of columns for years
    # We will extract from the balance sheet columns just to keep consistent
    available_years = df_balance.columns[1:]

    def row_to_dict(rdf):
        if rdf.empty:
            return {}
        # e.g. row df.iloc[0, 1:] -> the numeric cells, ignoring first column
        return dict(zip(available_years, rdf.iloc[0, 1:].values))

    # Convert each row to a dictionary
    total_assets = row_to_dict(total_assets_row)
    total_debt = row_to_dict(total_debt_row)
    total_equity = row_to_dict(equity_row)
    shares_issued = row_to_dict(shares_row)

    npi_dict = row_to_dict(npi_row)
    interest_exp_dict = row_to_dict(interest_expense_row)
    ebitda_dict = row_to_dict(ebitda_row)

    # Liabilities: if total_liab_row empty, sum Current + Non-current
    if total_liab_row.empty:
        curr_liab = find_balance_row("Current Liabilities")
        noncurr_liab = find_balance_row("Total Non Current Liabilities")
        curr_dict = row_to_dict(curr_liab)
        noncurr_dict = row_to_dict(noncurr_liab)
        total_liab = {}
        for col_yr in available_years:
            c_val = float(curr_dict.get(col_yr, 0) or 0)
            n_val = float(noncurr_dict.get(col_yr, 0) or 0)
            total_liab[col_yr] = c_val + n_val
    else:
        def row_to_dict2(rdf):
            return dict(zip(available_years, rdf.iloc[0, 1:].values))
        total_liab = row_to_dict2(total_liab_row)

    # From df_info
    market_cap = df_info["marketCap"].values[0] if "marketCap" in df_info.columns else None
    pub_pb = df_info["priceToBook"].values[0] if "priceToBook" in df_info.columns else None

    # Attempt Current Price
    try:
        
        # ticker = yf.Ticker(stock_code_yf)
        # cp_info = ticker.info
        # cprice = cp_info.get("currentPrice", None)
        
        # Previously using the yfinance, but now using self-built function 
        # Recrawl the data. Maybe will implement an update to change store the current price. 
        
        # 2025-08-13 TODO NOTE TO SELF - CRAWL AND STORE THE DATA ONCE 
        cprice = get_stock_price(stock_code_yf)
        print(f"Success!! Gathered current price for {stock_code_yf}: {cprice}")

    except:
        cprice = None
        
    # Get the dec_2024_price again
    # 31 December 2024 Stock Price via CSV
    dec_2024_price = None
    try:
        df_stock_price = pd.read_csv(stock_price_file)
        # Assuming the 5th column (index 4) is 'Close'.
        # If there's no header, you might need to specify header=None and use index directly.
        if 'Date' in df_stock_price.columns:
            # df_stock_price['Date'] = pd.to_datetime(df_stock_price['Date'], format='mixed').dt.date
            df_stock_price['Date'] = pd.to_datetime(df_stock_price['Date'], infer_datetime_format=True).dt.date

            target_date = pd.to_datetime('2024-12-31').date()
            row = df_stock_price[df_stock_price['Date'] == target_date]
            if not row.empty:
                closing_price = row.iloc[0, 4]  # Access the 5th column (index 4)
                rounded_price = round(closing_price, 2)
                print(f"The closing price for {target_date} is: {rounded_price}")
                dec_2024_price = rounded_price
            else:
                print(f"No data found for {target_date}")
                dec_2024_price = None
        else:
            # If there's no 'Date' column, we can't filter by date directly.
            # You'd need to know the row index for 2024-12-31 if that's the case.
            print("Warning: 'Date' column not found. Cannot filter by date.")
            dec_2024_price = None

    except FileNotFoundError:
        print(f"Error: File not found at {stock_price_file}")
        dec_2024_price = None
    except IndexError:
        print("Error: The 5th column (index 4) does not exist in the CSV.")
        dec_2024_price = None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        dec_2024_price = None

    # Dividends by year-col
    year_div_dict = {}
    if df_dividends is not None:
        for ycol in available_years:
            ycol_year = ycol[:4]
            year_div_dict[ycol] = df_dividends[
                df_dividends["Date"].astype(str).str.startswith(ycol_year)
            ]["Dividends"].sum()

    # For each year column, compute ratios
    for ycol in available_years:
        assets = float(total_assets.get(ycol, 0) or 0)
        debt = float(total_debt.get(ycol, 0) or 0)
        liab = float(total_liab.get(ycol, 0) or 0)
        eq = float(total_equity.get(ycol, 0) or 0)
        sh = float(shares_issued.get(ycol, 0) or 0)

        # Dividends (DPU) for that year
        dpu = float(year_div_dict.get(ycol, 0) or 0)

        # NPI, interest expense, property valuation from the Income file
        npi_val = float(npi_dict.get(ycol, 0) or 0)
        ie_val = float(interest_exp_dict.get(ycol, 0) or 0)
        ebitda_val = float(ebitda_dict.get(ycol, 0) or 0)

        # Basic items
        nav = (assets - liab) if (assets and liab) else None
        nav_unit = (nav / sh) if (nav and sh != 0) else None

        # Price-to-NAV Ratio
        price_to_nav = (cprice / nav_unit) 
        
        # Price-to-NAV Ratio using 31 Dec 2024 Price
        price_to_nav_2024 = ( dec_2024_price / nav_unit)
        
        # Gearing ratio (debt / assets)
        gear = (debt / assets) if assets != 0 else None
        # D/E
        d_e = (debt / eq) if eq != 0 else None
        # BVPS
        bvps = (eq / sh) if (eq and sh != 0) else None
        # Calc P/B
        calc_pb = (cprice / bvps) if (cprice and bvps and bvps != 0) else None
        
        # P/B 2024
        calc_pb_2024 = ( dec_2024_price / bvps )
        
        # Dividend Yield
        dyield = (dpu / cprice)*100 if (cprice and cprice != 0 and dpu) else None
        
        # Dividend Yield 2024
        dyield_2024 = (dpu / dec_2024_price)*100

        # Interest Coverage Ratio 
        # icr = (npi_val / ie_val) if (npi_val != 0 and ie_val != 0) else None
        icr = (ebitda_val / ie_val) if (ebitda_val != 0 and ie_val != 0) else None

        all_stocks_data.append([
            stock_code,
            reit["Name"],
            ycol,
            cprice,
            dec_2024_price,
            market_cap,
            nav,
            nav_unit,
            price_to_nav,
            price_to_nav_2024,
            bvps,
            pub_pb,
            calc_pb,
            calc_pb_2024,
            gear,
            d_e,
            dpu,
            dyield,
            dyield_2024,
            npi_val,
            ie_val,
            icr
        ])

# Build final_df (all years)
final_df = pd.DataFrame(all_stocks_data, columns=[
    "Stock Code","Company Name","Year","Current Price", "31 Dec 2024 Price", "Market Cap",
    "NAV","NAV (Per Unit)", "Price-to-NAV Ratio (Share Price / NAV Per Unit)", "Price-to-NAV Ratio (31 Dec 2024 Price)",
    "BVPS","Published P/B Ratio","Calculated P/B Ratio (Share Price / Book Value Per Share)", "Calculated P/B Ratio (31 Dec 2024 Price)",
    "Gearing Ratio (Total Debt/ Total Assets)","D/E Ratio (Debt / Shareholder Equity)","DPU","Dividend Yield", "Dividend Yield (31 Dec 2024 Price)",
    "Net Property Income", "Interest Expense", "Interest Coverage Ratio (EBITDA / Interest Expense)"
])

all_csv = "fundamentals/financial_ratios_all_stocks.csv"
final_df.to_csv(all_csv, index=False)
print(f"✅ All-year financial ratios saved to: {all_csv}")
```
